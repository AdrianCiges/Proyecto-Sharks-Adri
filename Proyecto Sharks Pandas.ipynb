### CORREGIMOS NOMBRES DE COLUMNAS

```python
Nombres_mal = list(data_sharks.columns)
Nombres_mal
```

```python
Nombres_bien = []

for nom in Nombres_mal:  #Hay una forma más fácil de hacerlo pero quería complicarme para aprender a pensar :D
    
    if nom[len(nom)-1] == " ":
        bien = nom.replace(" ","")
        bien2 = bien.capitalize()
        Nombres_bien.append(bien2)
    
    elif " " in nom and ":" in nom:
        bien = nom.replace(" ","_")
        bien2 = bien.replace(":","")
        bien3 = bien2.capitalize()
        Nombres_bien.append(bien3)
        
    elif " " in nom:
        bien = nom.replace(" ","_")
        bien2 = bien.capitalize()
        Nombres_bien.append(bien2)
        
    else:
        nom2 = nom.capitalize()
        Nombres_bien.append(nom2)
        
Nombres_bien
```

```python
len(Nombres_mal) == len(Nombres_bien) #Comprobar que no se ha cargado ninguna columna
```
